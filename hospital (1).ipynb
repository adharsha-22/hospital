```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hospital.csv')
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(15, 4))
```

```python
plt.subplot(1, 3, 1)
sns.countplot(data=df, x='Disease', palette='Set2')
plt.title('Disease Count')
plt.xticks(rotation=45)
plt.xlabel('Disease')
plt.ylabel('Number of Patients')
```

```python
plt.subplot(1, 3, 2)
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Male vs Female Patients')
```

```python
plt.subplot(1, 3, 3)
sns.barplot(data=df, x='Name', y='Bill amount', palette='viridis')
plt.title('Bill Amount per Patient')
plt.xticks(rotation=45)
plt.xlabel('Patient Name')
plt.ylabel('Bill Amount (Rs)')
```

```python
plt.tight_layout()
plt.show()
```

```python
print("=== SUMMARY ===")
print(f"Total Patients: {len(df)}")
print(f"Most Common Disease: {df['Disease'].mode()[0]}")
print(f"Average Bill: Rs. {df['Bill amount'].mean():.0f}")
```
